```python
import pandas as pd
df= pd.read_csv(r"C:\Users\ADMIN\Downloads\matches_2008-2024.csv")
df.head(10)
```

```python
df.info()
```

```python
#count how many matches team won

wins_by_team=df['winner'].value_counts()

wins_by_team
```

```python
import matplotlib.pyplot as plt
top_10= wins_by_team.head(10)
plt.bar(top_10.index,top_10.values)
plt.xticks(rotation=90)
plt.title('Top 10 Teams by Number of Wins')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.tight_layout()
plt.show()
```

```python
match_per_season= df['season'].value_counts()

match_per_season.sort_index()
```

```python
match_per_season_sorted= match_per_season.sort_index()
plt.bar(match_per_season_sorted.index, match_per_season_sorted.values)


plt.xticks(rotation=45, ha='right')
plt.title('Matches Played Per Season')
plt.xlabel('Season')
plt.ylabel('Matches')
plt.show()
```

```python
# Count Player of the Match awards
pom_counts = df['player_of_match'].value_counts()

# Top 10
pom_counts.head(10)
```

```python
top_10_pom = pom_counts.head(10)
plt.bar(top_10_pom.index, top_10_pom.values)
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Players by Player of the Match Awards')
plt.xlabel('Player')
plt.ylabel('Awards')
plt.tight_layout()
plt.show()
```

## Key Insights

- Mumbai Indians have the highest number of match wins across all seasons in this dataset, followed by Chennai Super Kings and Kolkata Knight Riders.
- In this dataset, the 2013 IPL season has the highest number of matches, with a total of 76 games played.
- In this dataset, AB de Villiers has won the most 'Player of the Match' awards, with a total of 25 awards.
